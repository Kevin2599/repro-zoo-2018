```python
# Concentrations
por = np.arange(31)/100

# Electric permittivity of constituents
ecalc = 7.5
ewater = 80.+0.045j

# Get aspect ratio
aratio_m = diemmod.aspratio(por, 'Markov')
aratio_s = diemmod.aspratio(por, 'Spherical')

# Calculate
crim = diemmod.eperm_crim(por, ecalc, ewater)
sigm_m = diemmod.eperm_ema(por, ecalc, ewater, aratio_m, crim)
sigm_s = diemmod.eperm_ema(por, ecalc, ewater, aratio_s, crim)
```

```python
plt.figure(figsize=(8, 3))

plt.subplot(121)
plt.title('Figure 2 A')
plt.xlabel('Porosity')
plt.ylabel(r"Real ($\varepsilon$)")
plt.plot(por, sigm_s.real, 'k:')
plt.plot(por, sigm_m.real, 'k--')
plt.plot(por, crim.real, 'k')
plt.axis([0, 0.3, 0, 30])

ax = plt.subplot(122)
plt.title('Figure 2 B')
plt.xlabel('Porosity')
plt.ylabel(r"Im ($\varepsilon$)")
plt.plot(por, sigm_s.imag, 'k:', label='EMA spherical pores')
plt.plot(por, sigm_m.imag, 'k--', label='EMA elliptic pores')
plt.plot(por, crim.imag, 'k-', label='CRIM')
plt.legend()
plt.axis([0, 0.3, 0, 0.0075])

plt.tight_layout()
plt.show()
```

**Fig. 2.** Real (A) and imaginary (B) parts of the effective dielectric permeability for calcite ($\varepsilon_c$ = 7.5) saturated with water as porosity functions at the frequency of 10 MHz. Curves 1 and 2 correspond to EMA model with spherical pores and pores with variable aspect ratios, respectively. Curve 3 corresponds to the semi-empirical CRIM model.

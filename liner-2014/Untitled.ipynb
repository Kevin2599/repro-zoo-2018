# Liner 2014

Liner, C (2014). Long-wave elastic attenuation produced by horizontal layering. _The Leading Edge_ **33** (6).

Much of the math has been coded up in [`bruges`](https://github.com/agile-geoscience/bruges). 

```python
from welly import Well

w = Well.from_las("R-39.las")
```

```python
dtp = w.data['DT4P']
rho = w.data['RHOB']
dts = w.data['DT4S']
depth = dtp.basis
dz = depth[1] - depth[0]

vp = 1e6 / dtp
vs = 1e6 / dts
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 2))
plt.plot(depth, vp)
```

## Figure 1

```python
import bruges

vp0, vs0, rho0 = bruges.rockphysics.backus(vp, vs, rho, lb=46, dz=dz)
```

```python
fig, axs = plt.subplots(nrows=2, figsize=(15, 4))

ax = axs[0]
ax.plot(depth, vp, 'k', lw=0.5)
ax.plot(depth, vp0, 'r')
ax.set_xlim(2800, 3300)
ax.set_ylim(3000, 5000)

ax = axs[1]
ax.plot(depth, vs, 'k', lw=0.5)
ax.plot(depth, vs0, 'r')
ax.set_xlim(2800, 3300)
```
